```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("validated_KADIRS_Tax_Compliance.csv")
```

```python
total_taxpayers = df["taxpayer_id"].nunique()

total_transactions = len(df)

total_assessed = df["amount_assessed"].sum()

total_revenue = df["amount_paid"].sum()

outstanding_balance = df["outstanding_balance"].sum()

average_compliance = df["compliance_score"].mean()

payment_rate = (df["payment_status"]=="Paid").mean()*100

filing_rate = (df["filing_status"]=="Filed").mean()*100

audit_rate = (df["audit_status"]=="Audited").mean()*100

high_risk = (df["risk_level"]=="High").sum()

medium_risk = (df["risk_level"]=="Medium").sum()

low_risk = (df["risk_level"]=="Low").sum()
```

```python
kpi_df = pd.DataFrame({

    "KPI":[
        "Total Taxpayers",
        "Total Transactions",
        "Total Tax Assessed",
        "Total Revenue Collected",
        "Outstanding Balance",
        "Average Compliance Score",
        "Payment Compliance Rate (%)",
        "Filing Compliance Rate (%)",
        "Audit Coverage (%)",
        "High Risk Taxpayers",
        "Medium Risk Taxpayers",
        "Low Risk Taxpayers"
    ],

    "Value":[
        total_taxpayers,
        total_transactions,
        total_assessed,
        total_revenue,
        outstanding_balance,
        round(average_compliance,2),
        round(payment_rate,2),
        round(filing_rate,2),
        round(audit_rate,2),
        high_risk,
        medium_risk,
        low_risk
    ]

})

kpi_df
```

```python
monthly_revenue = (
    df.groupby("tax_month")["amount_paid"]
      .sum()
      .reset_index()
)
```

```python
months = [
    "Jan","Feb","Mar","Apr","May","Jun",
    "Jul","Aug","Sep","Oct","Nov","Dec"
]

monthly_revenue["tax_month"] = pd.Categorical(
    monthly_revenue["tax_month"],
    categories=months,
    ordered=True
)

monthly_revenue = monthly_revenue.sort_values("tax_month")
```

```python
lga_revenue = (
    df.groupby("lga")["amount_paid"]
      .sum()
      .reset_index()
      .sort_values(
          by="amount_paid",
          ascending=False
      )
)
```

```python
sector_revenue = (
    df.groupby("business_sector")["amount_paid"]
      .sum()
      .reset_index()
      .sort_values(
          by="amount_paid",
          ascending=False
      )
)
```

```python
tax_revenue = (
    df.groupby("tax_type")["amount_paid"]
      .sum()
      .reset_index()
      .sort_values(
          by="amount_paid",
          ascending=False
      )
)
```

```python
tax_revenue = (
    df.groupby("tax_type")["amount_paid"]
      .sum()
      .reset_index()
      .sort_values(
          by="amount_paid",
          ascending=False
      )
)
```

```python
payment_summary = (
    df["payment_status"]
      .value_counts()
      .reset_index()
)

payment_summary.columns = [
    "Payment Status",
    "Count"
]
```

```python
filing_summary = (
    df["filing_status"]
      .value_counts()
      .reset_index()
)

filing_summary.columns = [
    "Filing Status",
    "Count"
]
```

```python
risk_summary = (
    df["risk_level"]
      .value_counts()
      .reset_index()
)

risk_summary.columns = [
    "Risk Level",
    "Count"
]
```

```python
audit_summary = (
    df["audit_status"]
      .value_counts()
      .reset_index()
)

audit_summary.columns = [
    "Audit Status",
    "Count"
]
```

```python
top_lga = (
    lga_revenue
    .head(10)
)
```

```python
top_sector = (
    sector_revenue
    .head(10)
)
```

```python
top_defaulters = (
    df.sort_values(
        by="outstanding_balance",
        ascending=False
    )
    .head(10)
)
```

```python
top_defaulters = top_defaulters[
    [
        "business_name",
        "lga",
        "amount_assessed",
        "amount_paid",
        "outstanding_balance"
    ]
]
```

```python
kpi_df.to_csv("kpi_summary.csv", index=False)

monthly_revenue.to_csv("monthly_revenue.csv", index=False)

lga_revenue.to_csv("lga_revenue.csv", index=False)

sector_revenue.to_csv("sector_revenue.csv", index=False)

tax_revenue.to_csv("tax_revenue.csv", index=False)

payment_summary.to_csv("payment_summary.csv", index=False)

filing_summary.to_csv("filing_summary.csv", index=False)

risk_summary.to_csv("risk_summary.csv", index=False)

audit_summary.to_csv("audit_summary.csv", index=False)

top_lga.to_csv("top_10_lga.csv", index=False)

top_sector.to_csv("top_10_sector.csv", index=False)

top_defaulters.to_csv("top_defaulters.csv", index=False)
```
